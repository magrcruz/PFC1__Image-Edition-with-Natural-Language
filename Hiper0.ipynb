```python
import matplotlib.pyplot as plt

# Cargar las im√°genes
img_source = plt.imread('dog.png')
img_edition = plt.imread('standing_dog/result/a sitting dog_seed111111_0.png')

# Crear una figura y dos subtramas
fig, (ax1, ax2) = plt.subplots(1, 2)

# Mostrar la imagen source en la primera subtrama
ax1.imshow(img_source)
ax1.set_title('Source')
ax1.axis('off')

# Mostrar la imagen edition en la segunda subtrama
ax2.imshow(img_edition)
ax2.set_title('Edition')
ax2.axis('off')

# Ajustar el dise√±o de las subtramas y la figura
plt.tight_layout()

# Mostrar la figura con las im√°genes
plt.show()
```
